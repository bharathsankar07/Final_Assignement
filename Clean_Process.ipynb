```python
%sql
SELECT 
  vehicle_type,
  COUNT(*) as total_rides,
  SUM(CASE WHEN booking_status = 'Completed' THEN 1 ELSE 0 END) as completed_rides,
  ROUND(
    SUM(CASE WHEN booking_status = 'Completed' THEN 1 ELSE 0 END) * 100.0 / COUNT(*), 
    2
  ) as success_rate
FROM uber_rides 
GROUP BY vehicle_type
ORDER BY success_rate DESC;
```

```python
%sql
SELECT 
  payment_method,
  COUNT(*) as count,
  ROUND(COUNT(*) * 100.0 / (SELECT COUNT(*) FROM uber_rides), 2) as percentage
FROM uber_rides 
WHERE payment_method IS NOT NULL
GROUP BY payment_method
ORDER BY count DESC;
```

```python
%sql
SELECT 
  pickup_location,
  COUNT(*) as ride_count
FROM uber_rides 
WHERE booking_status = 'Completed'
GROUP BY pickup_location
ORDER BY ride_count DESC
LIMIT 10;
```
